```python
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets, models
import torch.nn.functional as F
from sklearn.metrics import f1_score

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Path to dataset
data_dir = "wec-intelligence-sig-2024-recruitment-task-cv/train_images/train_images"
```

```python
def visualize_sample_images(data_dir, num_samples=4):
    real_dir = os.path.join(data_dir, 'real_train')
    fake_dir = os.path.join(data_dir, 'fake_train')
    
    real_images = os.listdir(real_dir)[:num_samples]  # Limit to `num_samples`
    fake_images = os.listdir(fake_dir)[:num_samples]  # Limit to `num_samples`

    fig, axes = plt.subplots(2, num_samples, figsize=(12, 6))
    
    # Plot real images
    for i, img_name in enumerate(real_images):
        img = Image.open(os.path.join(real_dir, img_name))
        axes[0, i].imshow(img)
        axes[0, i].set_title('Real')
        axes[0, i].axis('off')
    
    # Plot fake images
    for i, img_name in enumerate(fake_images):
        img = Image.open(os.path.join(fake_dir, img_name))
        axes[1, i].imshow(img)
        axes[1, i].set_title('Fake')
        axes[1, i].axis('off')

    plt.show()

# Call function to display few images (e.g., 3 from each class)
visualize_sample_images(data_dir, num_samples=3)
```

```python
def class_distribution(data_dir):
    real_count = len(os.listdir(os.path.join(data_dir, 'real_train')))
    fake_count = len(os.listdir(os.path.join(data_dir, 'fake_train')))
    
    labels = ['Real', 'Fake']
    counts = [real_count, fake_count]

    plt.bar(labels, counts, color=['blue', 'red'])
    plt.title('Class Distribution')
    plt.show()

class_distribution(data_dir)
```

```python
class DeepfakeDataset(Dataset):
    def __init__(self, data_dir, transform=None):
        self.real_dir = os.path.join(data_dir, 'real_train')
        self.fake_dir = os.path.join(data_dir, 'fake_train')
        self.real_images = os.listdir(self.real_dir)
        self.fake_images = os.listdir(self.fake_dir)
        self.transform = transform

    def __len__(self):
        return len(self.real_images) + len(self.fake_images)

    def __getitem__(self, idx):
        if idx < len(self.real_images):
            img_path = os.path.join(self.real_dir, self.real_images[idx])
            label = 0  # Real
        else:
            img_path = os.path.join(self.fake_dir, self.fake_images[idx - len(self.real_images)])
            label = 1  # Fake

        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        return image, label

# Data augmentation and transformation
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
])

# Load dataset
dataset = DeepfakeDataset(data_dir, transform=transform)
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# DataLoader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
class DeepfakeCNN(nn.Module):
    def __init__(self):
        super(DeepfakeCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(128 * 8 * 8, 512)
        self.fc2 = nn.Linear(512, 1)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(self.bn1(F.relu(self.conv1(x))))
        x = self.pool(self.bn2(F.relu(self.conv2(x))))
        x = self.pool(self.bn3(F.relu(self.conv3(x))))
        x = x.view(-1, 128 * 8 * 8)  # Flatten
        x = self.dropout(F.relu(self.fc1(x)))
        x = torch.sigmoid(self.fc2(x))
        return x

model = DeepfakeCNN().to(device)
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
```

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=15):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    
    for epoch in range(num_epochs):
        model.train()  # Set the model to training mode
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.float().to(device)
            
            optimizer.zero_grad()
            
            outputs = model(inputs)
            loss = criterion(outputs.squeeze(), labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()

        # Validation phase
        model.eval()  # Set the model to evaluation mode
        val_loss = 0.0
        correct = 0
        total = 0
        all_preds = []
        all_labels = []
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.float().to(device)
                
                outputs = model(inputs)
                loss = criterion(outputs.squeeze(), labels)
                val_loss += loss.item()
                
                # Convert sigmoid outputs to binary predictions
                preds = (outputs.squeeze() > 0.5).float()
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                
                correct += (preds == labels).sum().item()
                total += labels.size(0)
        
        # Calculate metrics
        val_accuracy = correct / total * 100
        f1 = f1_score(all_labels, all_preds)  # F1 score calculation
        
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader):.4f}, '
              f'Val Loss: {val_loss/len(val_loader):.4f}, Val Accuracy: {val_accuracy:.2f}%, '
              f'F1 Score: {f1:.4f}')

train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=15)
```

```python
from sklearn.metrics import classification_report

def evaluate_model(model, val_loader):
    model.eval()
    y_true = []
    y_pred = []
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            predicted = (outputs.squeeze() > 0.5).float()
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

    print(classification_report(y_true, y_pred, target_names=['Real', 'Fake']))

evaluate_model(model, val_loader)
```

```python
import os
import pandas as pd
import torch
from torchvision import transforms
from PIL import Image

# Define the transformation
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # Adjust to the input size of your model
    transforms.ToTensor(),
])

# Function to predict images
def predict_images(model, test_dir):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()  # Set the model to evaluation mode
    model.to(device)
    
    predictions = []

    # Iterate through test images in the directory
    for filename in os.listdir(test_dir):
        if filename.startswith('test_image_') and filename.endswith(('.png', '.jpg', '.jpeg')):  # Adjust based on your file types
            image_id = filename.split('_')[-1].split('.')[0]  # Extract image ID from the filename
            image_path = os.path.join(test_dir, filename)
            image = Image.open(image_path).convert('RGB')  # Load and convert image to RGB
            image = transform(image).unsqueeze(0).to(device)  # Preprocess and add batch dimension
            
            with torch.no_grad():
                output = model(image)  # Forward pass
                prediction = (output.squeeze() > 0.5).float().cpu().numpy()  # Convert to binary prediction
                predictions.append((image_id, int(prediction)))

    # Create a DataFrame
    df = pd.DataFrame(predictions, columns=['ID', 'TARGET'])
    
    # Display the DataFrame
    return df

# Assuming you have already trained your model and it's named 'model'
# Call the function
test_directory = 'wec-intelligence-sig-2024-recruitment-task-cv/test_images/test_images'  # Update with your test images path
output=predict_images(model, test_directory)
```
